```python
import pandas as pd
df1=pd.read_csv('accidents_2005_to_2007.csv',low_memory=False)
df2=pd.read_csv('accidents_2009_to_2011.csv',low_memory=False)
df3=pd.read_csv('accidents_2012_to_2014.csv',low_memory=False)
df1.describe()
```

```python
df=pd.concat([df1,df2,df3],ignore_index=True)
```

```python
df.describe()
```

```python
df.isnull().sum().sort_values(ascending=False)
```

```python
mean_vehicles=df.groupby('Accident_Severity')['Number_of_Vehicles'].mean()
mean_vehicles
```

Anlaşılacağı üzere daha ölümcül kazalarda araç sayıları daha fazla oluyor

```python
import seaborn as sns
import matplotlib.pyplot as plt
mean_vehicles.plot(kind='bar')
plt.xlabel('Kaza Ciddiyeti')
plt.ylabel('Ortalama Araç Sayısı')
plt.title("Kaza Ciddiyeti ve Ortalama Araç Sayısı (Bar Plot)")
plt.xticks(rotation=0)
plt.show()
```

```python
grup1=df[df["Accident_Severity"]==1]["Number_of_Vehicles"]
grup2=df[df["Accident_Severity"]==2]["Number_of_Vehicles"]
grup3=df[df["Accident_Severity"]==3]["Number_of_Vehicles"]
```

```python
from scipy.stats import f_oneway
f_stat,p_val=f_oneway(grup1,grup2,grup3)
print('f_statics',f_stat)
print('p_degeri',p_val)
```

```python
from statsmodels.stats.multicomp import pairwise_tukeyhsd
tukey=pairwise_tukeyhsd(endog=df1['Number_of_Vehicles'],
                  groups=df1['Accident_Severity'],
                  alpha=0.05)
print(tukey)
```

Kazanın ciddiyeti arttıkça kazaya karışan araç sayısı da artıyor.



p value “Bu fark tesadüfen mi oluştu, yoksa gerçekten anlamlı mı?” demek 


p-adj < 0.05 → ✅ Fark anlamlı (reject = True olur)

p-adj ≥ 0.05 → ❌ Fark anlamlı değil (rastlantı olabilir)

"Yaya geçidinde insan kontrolü olması, kaza ciddiyetini etkiler mi?"


```python
df.groupby('Pedestrian_Crossing-Human_Control')['Accident_Severity'].mean()
```

```python
df['Human_Control_Code'] = df['Pedestrian_Crossing-Human_Control'].map({
    'None within 50 metres': 'Yok',
    'Control by other authorised person': 'Görevli',
    'Control by school crossing patrol': 'Okul Görevlisi'
})
```

```python
grup1=df[df['Human_Control_Code']=='Yok']['Accident_Severity']
grup2=df[df['Human_Control_Code']=='Görevli']['Accident_Severity']
grup3=df[df['Human_Control_Code']=='Okul Görevlisi']['Accident_Severity']
f_stat, p_val = f_oneway(grup1, grup2, grup3)
print("F-statistic:", f_stat)
print("p-değeri:", p_val)
```

Buradan görüyoruz ki insan kontrolü olması kaza ciddiyetini büyük bir oranda etkiliyor.
